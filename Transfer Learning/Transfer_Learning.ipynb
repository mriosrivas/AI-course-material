```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
```

# Prepare train/val datasets

```python
# Image processing functions
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Xception model
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.applications.xception import decode_predictions

# Neural Network layers
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model

# Checkpoint
from tensorflow.keras.callbacks import ModelCheckpoint

# Optimizer
from tensorflow.keras.optimizers import Adam
# Loss function
from tensorflow.keras.losses import CategoricalCrossentropy
# Dropout
from tensorflow.keras.layers import Dropout
```

## Data preparation

```python
train_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
train_ds = train_gen.flow_from_directory(
    'clothing-dataset-small/train/',
    target_size=(150,150),
    batch_size=32)
```

```python
val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
val_ds = val_gen.flow_from_directory(
    'clothing-dataset-small/validation/',
    target_size=(150,150),
    batch_size=32,
    shuffle=False)
```

## Transfer Learning

```python
def make_model(learning_rate = 0.01):
    # Define image input shape
    input_shape = (150,150,3)
    
    # Load base model
    base_model = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Set model as not trainable in provided layers
    base_model.trainable = False
    
    # Set input tensor
    inputs = Input(shape=input_shape)
    
    # Add input tensor to base model
    # Note: output shape [None, 5, 5, 2048]
    base = base_model(inputs, training=False) 
    
    # Use global average pooling to last layer.
    # It makes shape [None, 5, 5, 2048] into [None, 2048]
    # Note: output shape [None, 2048]
    vector = GlobalAveragePooling2D()(base)

    # Use a 10 neuron dense layer
    # Note: output shape [None, 10]
    outputs = Dense(10)(vector)
    
    # Define input/output model
    model = Model(inputs, outputs)
    
    
    # Use ADAM optimizer
    optimizer = Adam(learning_rate=learning_rate)
    
    # Define loss function
    loss = CategoricalCrossentropy(from_logits=True)
    
    # Compile model
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

```python
model = make_model(learning_rate=0.01)
```

```python
history = model.fit(train_ds, epochs=10, validation_data=val_ds)
```

```python
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

plt.rcParams['figure.figsize'] = (7,5)
plt.plot(train_accuracy, label='train')
plt.plot(val_accuracy, label='val')
plt.title('Accuracy vs. Epoch')
plt.grid('on')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

## Adjusting the Learning Rate

```python
scores = {}

for lr in [0.1, 0.01, 0.001, 0.0001]:
    model = make_model(learning_rate=lr)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[lr] = history.history
```

```python
plt.rcParams['figure.figsize'] = (7,5)

for key in scores.keys():
    #train_accuracy = scores[key]['accuracy']
    val_accuracy = scores[key]['val_accuracy']
    #plt.plot(train_accuracy, label=f'train = {key}')
    plt.plot(val_accuracy, label=f'val = {key}')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. Epoch')
    plt.grid('on')
    plt.legend()
plt.show()
```

```python
# Best learning rate
learning_rate_best = 0.001
```

## Adding more Layers

```python
def make_model_deeper(learning_rate = 0.001, size=100):
    # Define image input shape
    input_shape = (150,150,3)
    
    # Load base model
    base_model = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Set model as not trainable in provided layers
    base_model.trainable = False
    
    # Set input tensor
    inputs = Input(shape=input_shape)
    
    # Add input tensor to base model
    # Note: output shape [None, 5, 5, 2048]
    base = base_model(inputs, training=False) 
    
    # Use global average pooling to last layer.
    # It makes shape [None, 5, 5, 2048] into [None, 2048]
    # Note: output shape [None, 2048]
    vector = GlobalAveragePooling2D()(base)
    
    # Added dense layer
    # Note: output shape [None, size]
    inner = Dense(size, activation='relu')(vector)

    # Use a 10 neuron dense layer
    # Note: output shape [None, 10]
    outputs = Dense(10)(inner)
    
    # Define input/output model
    model = Model(inputs, outputs)
    
    
    # Use ADAM optimizer
    optimizer = Adam(learning_rate=learning_rate)
    
    # Define loss function
    loss = CategoricalCrossentropy(from_logits=True)
    
    # Compile model
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

```python
scores = {}

for size in [100, 500, 1000]:
    model = make_model_deeper(learning_rate=learning_rate_best, size=size)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[size] = history.history
```

```python
plt.rcParams['figure.figsize'] = (7,5)

for key in scores.keys():
    train_accuracy = scores[key]['accuracy']
    val_accuracy = scores[key]['val_accuracy']
    plt.plot(train_accuracy, label=f'train = {key}', linestyle='--')
    plt.plot(val_accuracy, label=f'val = {key}')
    plt.grid('on')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. Epoch')
    plt.legend()
plt.show()
```

```python
# Best size
size_best = 100
```

## Dropout and Regularization

```python
def make_model_dropout(learning_rate = 0.001, size=100, droprate=0.5):
    # Define image input shape
    input_shape = (150,150,3)
    
    # Load base model
    base_model = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Set model as not trainable in provided layers
    base_model.trainable = False
    
    # Set input tensor
    inputs = Input(shape=input_shape)
    
    # Add input tensor to base model
    # Note: output shape [None, 5, 5, 2048]
    base = base_model(inputs, training=False) 
    
    # Use global average pooling to last layer.
    # It makes shape [None, 5, 5, 2048] into [None, 2048]
    # Note: output shape [None, 2048]
    vector = GlobalAveragePooling2D()(base)
    
    # Added dense layer
    # Note: output shape [None, size]
    inner = Dense(size, activation='relu')(vector)
    
    # Added dropout layer
    # Note: output shape [None, size]
    drop = Dropout(droprate)(inner)

    # Use a 10 neuron dense layer
    # Note: output shape [None, 10]
    outputs = Dense(10)(drop)
    
    # Define input/output model
    model = Model(inputs, outputs)
    
    
    # Use ADAM optimizer
    optimizer = Adam(learning_rate=learning_rate)
    
    # Define loss function
    loss = CategoricalCrossentropy(from_logits=True)
    
    # Compile model
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

```python
scores = {}

for droprate in [0.0, 0.2, 0.5, 0.8]:
    model = make_model_dropout(learning_rate=learning_rate_best, size=size_best, droprate=droprate)
    history = model.fit(train_ds, epochs=10, validation_data=val_ds)
    scores[droprate] = history.history
```

```python
plt.rcParams['figure.figsize'] = (7,5)

for key in scores.keys():
    train_accuracy = scores[key]['accuracy']
    val_accuracy = scores[key]['val_accuracy']
    plt.plot(train_accuracy, label=f'train droprate = {key}', linestyle='--')
    plt.plot(val_accuracy, label=f'val droprate = {key}')
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs. Epoch')
    plt.grid('on')
    plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (7,5)

drop = 0.8
train_accuracy = scores[drop]['accuracy']
val_accuracy = scores[drop]['val_accuracy']
plt.plot(train_accuracy, label=f'train droprate = {key}')
plt.plot(val_accuracy, label=f'val droprate = {key}')
plt.title('Accuracy vs. Epoch')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.grid('on')
plt.legend()
plt.show()
```

## Training a larger network

```python
image_size = 299
```

```python
def make_model_larger(image_size=150, learning_rate = 0.001, size=100, droprate=0.5):
    # Define image input shape
    input_shape = (image_size,image_size,3)
    
    # Load base model
    base_model = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Set model as not trainable in provided layers
    base_model.trainable = False
    
    # Set input tensor
    inputs = Input(shape=input_shape)
    
    # Add input tensor to base model
    # Note: output shape [None, 5, 5, 2048]
    base = base_model(inputs, training=False) 
    
    # Use global average pooling to last layer.
    # It makes shape [None, 5, 5, 2048] into [None, 2048]
    # Note: output shape [None, 2048]
    vector = GlobalAveragePooling2D()(base)
    
    # Added dense layer
    # Note: output shape [None, size]
    inner = Dense(size, activation='relu')(vector)
    
    # Added dropout layer
    # Note: output shape [None, size]
    drop = Dropout(droprate)(inner)

    # Use a 10 neuron dense layer
    # Note: output shape [None, 10]
    outputs = Dense(10)(drop)
    
    # Define input/output model
    model = Model(inputs, outputs)
    
    
    # Use ADAM optimizer
    optimizer = Adam(learning_rate=learning_rate)
    
    # Define loss function
    loss = CategoricalCrossentropy(from_logits=True)
    
    # Compile model
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

```python
train_gen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    shear_range=10.0,
    zoom_range=0.6,
    vertical_flip=True)

train_ds = train_gen.flow_from_directory(
    'clothing-dataset-small/train/',
    target_size=(image_size, image_size),
    batch_size=32)

val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)

val_ds = val_gen.flow_from_directory(
    'clothing-dataset-small/validation/',
    target_size=(image_size, image_size),
    batch_size=32,
    shuffle=False)
```

```python
checkpoint = ModelCheckpoint(
    'checkpoints/xception_v4_{epoch:02d}_{val_accuracy:.3f}.h5',
    save_best_only=True, 
    monitor='val_accuracy', 
    mode='max')
```

```python
model = make_model_larger(image_size=image_size, 
                          learning_rate=learning_rate_best, 
                          size=size_best, 
                          droprate=droprate_best)

history = model.fit(train_ds, 
                    epochs=15, 
                    validation_data=val_ds, 
                    callbacks=[checkpoint])

scores = history.history
```
