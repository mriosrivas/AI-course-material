```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing import image
```

```python
path = 'clothing-dataset-small/train/shirt/'
file = '2a774d80-c007-43a7-9c38-85e441274cfc.jpg'
```

```python
img = load_img(path+file, target_size=(299,299))
img
```

## Xception Model
https://keras.io/api/applications/xception/

```python
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.applications.xception import decode_predictions
```

```python
model_xception = Xception(weights='imagenet', input_shape=(299, 299, 3))
```

```python
# Create array of [1, 299, 299, 3]
# X = preprocess_input(np.array([np.array(img)]))
# Example with two images
X = preprocess_input(np.array([np.array(img), np.array(img)]))
```

```python
y_pred = model_xception.predict(X)
```

```python
decode_predictions(y_pred, top=3)
```
