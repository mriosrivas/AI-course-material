```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('xception_v4_10_0.862.h5')
```

```python
classes = ['dress', 'hat', 'longsleeve', 'outwear', 'pants', 'shirt', 'shoes', 'shorts', 'skirt', 't-shirt']
```

```python
from tensorflow.keras.applications.xception import preprocess_input
```

```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
img = load_img('shorts.jpg', target_size=(299,299))
img
```

```python
x = np.array(img)
x = np.array([x])
```

```python
X = preprocess_input(x)
```

```python
pred = model.predict(X)[0]
```

```python
dict(zip(classes, pred))
```

## Change data preprocessing to avoid usage of Tensorflow

```python
from PIL import Image
```

```python
with Image.open('shorts.jpg') as img:
    img = img.resize((299, 299), Image.NEAREST)
```

```python
def preprocess_lite(X):
    #X /= 127.5
    #X -= 1.
    return np.float32((X/127.5) - 1)
```

```python
X = np.array(img)
X = np.array([X])
X.shape
```

```python
X = preprocess_lite(X)
```

## Convert Keras to TF-Lite

```python
import tensorflow.lite as tflite
```

```python
converter = tflite.TFLiteConverter.from_keras_model(model)

tflite_model = converter.convert()

with open('xception.tflite', 'wb') as file:
    file.write(tflite_model)
```

```python
# Load model to interpreter
interpreter = tflite.Interpreter(model_path='xception.tflite')
# Load weights
interpreter.allocate_tensors()
```

```python
# Find input pointer
input_index = interpreter.get_input_details()[0]['index']
```

```python
# Find output pointer
output_index = interpreter.get_output_details()[0]['index']
```

```python
# Load data into input
interpreter.set_tensor(input_index, X)
```

```python
# Run inference
interpreter.invoke()
```

```python
# Load output into memory
lite_pred = interpreter.get_tensor(output_index)
```

```python
dict(zip(classes, lite_pred[0]))
```
